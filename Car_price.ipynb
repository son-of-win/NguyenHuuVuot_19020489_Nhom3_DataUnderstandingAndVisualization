```python
import pandas as pd
import matplotlib.pyplot as plt
```

Khai báo các thư viện sẽ dùng

```python
df = pd.read_csv('USA_Cars.csv')
```

Lấy dữ liệu từ Data

```python
price = df['price'].value_counts()  # dict type
year = df['year'].value_counts()    # dict type
color = df['color'].value_counts()  # dict type
brand = df['brand'].value_counts()
model = df['model'].value_counts()
```

# Số lượng xe đc sản xuất theo năm

```python
year_key = year.keys()
year_value = [year[x] for x in year_key]
plt.xlabel("year")
plt.ylabel('Car')
plt.bar(year_key,year_value)
plt.show()
```

# Số lượng màu đc các hãng sản xuất

```python
color_key = color.keys()
color_value = [color[x] for x in color_key]
plt.xlabel("Color")
plt.ylabel('Car')

plt.barh(color_key,color_value)
plt.tick_params(axis='y',which='major',labelsize=4)
plt.show()
```

 # Số lượng xe của các hãng và giá trung bình


```python
brand_key = brand.keys()
brand_value = [brand[x] for x in brand_key]
```

```python
price_average = []
for x in brand_key:
	price_sum = df.loc[df['brand'] == x,'price']
	price_average.append((price_sum.sum()/price_sum.count()))
plt.plot(brand_key,price_average, label = 'average_price')
plt.plot(brand_key, brand_value,color='red',label = 'color')
plt.legend()
plt.tick_params(axis='x', which='major', labelsize=4)  # chỉnh cỡ chữ trên biểu đồ
plt.show()
```

# Độ chiếm lĩnh thị phần của các hãng

```python
total = sum(brand_value)
percent = [round((brand[x] / total)*100,2) for x in brand_key]
plt.pie(percent, labels=brand_key, autopct='%1.2f%%')
plt.show()
```
